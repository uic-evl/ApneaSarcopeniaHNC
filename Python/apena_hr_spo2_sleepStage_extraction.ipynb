```python
import pandas as pd
import numpy as np
import pickle
import json
```

```python
with open('processed_patients.pickle','rb') as f:
    data = pickle.load(f)
data = data.reset_index().drop_duplicates('patient_id').set_index('patient_id')
data.head()
```

```python
temp = data.copy()

temp.apnea = temp.apnea.astype(int)
temp.SpO2 = temp.SpO2.apply(lambda x: x.tolist())
temp.HR = temp.HR.apply(lambda x: x.tolist())
temp.Stage = temp.Stage.apply(lambda x: x.tolist())

temp = temp[['apnea', 'SpO2', 'HR', 'Stage']].to_dict(orient='records')

# print(temp)

with open('spo2_apnea_hr_stage_records.json', 'w', encoding='utf-8') as f: 
    json.dump(temp, f, ensure_ascii=False)
```
